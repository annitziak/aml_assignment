```python
import re

#"geographic distribution"
#body mass?
traits_list = ["Species ID","conservation status","locomotion","body mass (g)", "behavioral circadian rhythm","habitat","reproduction","trophic guild"]

traits_df = traits[traits_list]

def clean_text(text):
    text = str(text)  # Convert to string to avoid issues with non-string types
    if pd.isna(text) or 'URI' not in text:  # Leave NaN or cells without 'URI' as they are
        return text.strip("()").replace("'", "")  # Remove parentheses and single quotes if present
    # Use regex to find everything before 'URI' and remove parentheses and single quotes
    match = re.search(r'^(.*?)(?=URI)', text)
    cleaned_text = match.group(1).strip() if match else text
    return cleaned_text.strip("()").replace("'", "")  # Remove parentheses and single quotes from cleaned text

# Apply the function to each column in traits_list
for column in traits_list:
    traits_df[column] = traits_df[column].apply(clean_text)

    
# average for mass
traits_df["body mass (g)"] = traits_df["body mass (g)"].apply(
    lambda x: np.mean([float(i) for i in str(x).split(',') if i.strip()]) if pd.notna(x) else x)

#group better
traits_df["conservation status"] = traits_df["conservation status"].replace(
    {"cites appendix i": "vulnerable (trade)",
     "cites appendix ii": "vulnerable (trade)",
     "cites appendix iii": "vulnerable (trade)"}
)

#need to merge
```

```python
traits_df["Species ID"] = pd.to_numeric(traits_df["Species ID"], errors='coerce')
animals_df["id"] = pd.to_numeric(animals_df["id"], errors='coerce')

# Debugging: Display unique values to ensure compatibility
print("Unique values in traits_df['Species ID']:\n", traits_df["Species ID"].unique())
print("\nUnique values in animals_df['id']:\n", animals_df["id"].unique())

# Perform the merge
merged_df = pd.merge(traits_df, animals_df[['id', 'class']], left_on='Species ID', right_on='id', how='left')

# Drop the 'id' column if not needed
merged_df = merged_df.drop(columns=['id'])

# Check the result
print("\nMerged DataFrame:\n", merged_df.head())
```
